### Step 1. Import the necessary libraries

### Step 3. Assign it to a variable called crime.

```python
crime=pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv")
crime
```

### Step 4. What is the type of the columns?

```python
crime.dtypes
```

##### Have you noticed that the type of Year is int64. But pandas has a different type to work with Time Series. Let's see it now.

### Step 5. Convert the type of the column Year to datetime64

```python
crime['Year'] = pd.to_datetime(crime['Year'])
crime.Year.dtype
```

### Step 6. Set the Year column as the index of the dataframe

```python
present=crime.set_index("Year")
present
```

### Step 7. Delete the Total column

```python
present = present.drop(columns='Total')
```

### Step 8. Group the year by decades and sum the values

#### Pay attention to the Population column number, summing this column is a mistake, use max for Population instead

```python
present.reset_index()
decade= ['60s','60s','60s','60s','60s','60s','60s','60s','60s','60s','70s','70s','70s','70s','70s','70s','70s','70s','70s','70s','80s','80s','80s','80s','80s','80s','80s','80s','80s','80s','90s','90s','90s','90s','90s','90s','90s','90s','90s','90s','200s','200s','200s','200s','200s','200s','200s','200s','200s','200s','201s','201s','201s','201s','201s']
present['decade'] = decade

res = present.groupby('decade', as_index=False,).sum()
print(res)
```

### Step 9. What is the most dangerous decade category wise?

```python
res.max(axis = 0)
```

### Step 5. Assign the columns as below:

The attributes are (donated by Riccardo Leardi, riclea '@' anchem.unige.it):  
1) alcohol  
2) malic_acid  
3) alcalinity_of_ash

7) hue 

```python
wine = wine.rename(columns={wine.columns[0]: 'alcohol'})
wine = wine.rename(columns={wine.columns[1]: 'malic_acid'})
wine = wine.rename(columns={wine.columns[2]: 'alcalinity_of_ash'})
wine = wine.rename(columns={wine.columns[3]: 'magnesium'})
wine = wine.rename(columns={wine.columns[4]: 'flavanoids'})
wine = wine.rename(columns={wine.columns[5]: 'proanthocyanins'})
wine = wine.rename(columns={wine.columns[6]: 'hue'})
wine
```

### Step 6. Set the values of the first 3 rows from alcohol as NaN

```python
wine.loc[:2,"alcohol"]=np.NaN
wine
```

### Step 7. Now set the value of the third and fourth of magnesium as NaN

```python
wine.loc[2:3,"magnesium"]=np.NaN
wine
```

### Step 8. Fill the value of NaN with the number 10 in alcohol and 100 in magnesium

```python
wine=wine.fillna({
    'alcohol':10,
    'magnesium':100
})
wine
```

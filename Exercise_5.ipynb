### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv). 

### Step 3. Assign it to a variable called 

```python
address = pd.read_csv (r'https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv')
address
```

### Step 4.  What is the frequency of the dataset (Yearly, Daily etc)?

```python
count = address.sort_values(by=['Date'])
print(count)
```

### Step 5. Set the column Date as the index.

```python
address.set_index('Date', inplace=True)
address
```

### Step 6. What is the type of the index?

```python
address.info()
print(address.index.dtype)
```

### Step 7. Set the index to a DatetimeIndex type

```python
address.index=pd.to_datetime(address.index)
print(address.index.dtype)
```

### Step 8.  Change the frequency to monthly, sum the values and assign it to monthly.

```python
address=address.resample('1M').sum()
address
```

### Step 9. You will notice that it filled the dataFrame with months that don't have any data with 0s. Replace 0s with NaN and then drop these rows.

```python
address=address.replace(0, np.nan)
print(address)
address=address.dropna()
print(address)
```

### Step 10. Good, now we have the monthly data. Now change the frequency to year.

```python
address=address.resample('1Y').sum()
address
```
